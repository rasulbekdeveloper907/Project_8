unique_km = sorted(df['kilometer'].unique())

## Continuous qiymat yaratish uchun har bir original qiymat oraligâ€˜ini belgilash:

for i, km in enumerate(unique_km):
    if i == len(unique_km) - 1:
        upper = km + 10000  # Oxirgi qiymat uchun yuqori limit

    mask = df['kilometer'] == km
